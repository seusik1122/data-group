## in 연산자를 이용한 구로구의 초미세먼지 데이터 출력

```python
f = open('dust.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)  

for row in data:
    if '구로구' in row[2]:
        print(f"날짜: {row[0]}, 초미세먼지: {row[8]}㎍/㎥")

f.close()
```

## 구로구 미세먼지 데이터 2달 간격으로 시각화

```python
import csv
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from datetime import datetime

font_path = 'C:/Windows/Fonts/malgun.ttf'
font_prop = fm.FontProperties(fname=font_path)

미세먼지_데이터 = []
날짜_데이터 = []

f = open('dust.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

for row in data:
    if '구로구' in row[2]:
        date = row[0]
        date_obj = datetime.strptime(date, '%Y%m%d')
        
        if date_obj.day == 1: # 매월 1일 데이터만 선택을 한다.
            미세먼지_데이터.append(float(row[7]))
            날짜_데이터.append(date_obj)

f.close()

날짜_데이터, 미세먼지_데이터 = zip(*sorted(zip(날짜_데이터, 미세먼지_데이터))) # 날짜순으로 데이터 정렬한다.

plt.figure(figsize=(12, 6))
plt.plot(날짜_데이터, 미세먼지_데이터, color='orange', marker='o', linewidth=2, markersize=8)
plt.ylim(0, max(미세먼지_데이터) + 20)
plt.title('구로구의 2달별 초미세먼지 농도 변화', fontproperties=font_prop)
plt.xlabel('날짜', fontproperties=font_prop)
plt.ylabel('초미세먼지 농도 (㎍/㎥)', fontproperties=font_prop)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 구로구의 초미세먼지 및 미세먼지 비교 막대 그래프

```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')

미세먼지_데이터 = []  
초미세먼지_데이터 = []
날짜_데이터 = []

f = open('dust.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

for row in data:
   if '구로구' in row[2]:
       날짜 = row[0]
       날짜_객체 = datetime.strptime(날짜, '%Y%m%d')

       if 날짜_객체.day == 1:
           미세먼지_데이터.append(float(row[7]))
           초미세먼지_데이터.append(float(row[8]))
           날짜_데이터.append(날짜_객체.strftime('%Y-%m'))

f.close()

날짜_데이터, 미세먼지_데이터, 초미세먼지_데이터 = zip(*sorted(zip(날짜_데이터, 미세먼지_데이터, 초미세먼지_데이터)))

plt.figure(figsize=(12, 6))
막대_너비 = 0.35
투명도 = 0.8

인덱스 = range(len(날짜_데이터))

plt.bar(인덱스, 미세먼지_데이터, 막대_너비, alpha=투명도, color='b', label='미세먼지')
plt.bar([x + 막대_너비 for x in 인덱스], 초미세먼지_데이터, 막대_너비, alpha=투명도, color='r', label='초미세먼지')

plt.xlabel('날짜')
plt.ylabel('먼지 농도 (㎍/㎥)')
plt.title('구로구의 월별 미세먼지와 초미세먼지 농도 비교')
plt.xticks([x + 막대_너비/2 for x in 인덱스], 날짜_데이터, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## plt.subplots()을 사용한 미세먼지, 초미세먼지 비교 그래프

```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')

미세먼지_데이터 = []
초미세먼지_데이터 = []
날짜_데이터 = []

f = open('dust.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

for row in data:
   if '구로구' in row[2]:
       날짜 = row[0]
       날짜_객체 = datetime.strptime(날짜, '%Y%m%d')

       if 날짜_객체.day == 1:
           미세먼지_데이터.append(float(row[7]))
           초미세먼지_데이터.append(float(row[8]))
           날짜_데이터.append(날짜_객체.strftime('%Y-%m'))

f.close()

날짜_데이터.reverse()
미세먼지_데이터.reverse()
초미세먼지_데이터.reverse()

fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(12, 6))
fig.subplots_adjust(wspace=0)

axes[0].barh(날짜_데이터, 미세먼지_데이터, align='center', color='skyblue')
axes[0].set(title='미세먼지')
axes[0].invert_xaxis()
axes[0].set_yticks([])
axes[0].grid(True, axis='x')

axes[1].barh(날짜_데이터, 초미세먼지_데이터, align='center', color='lightpink')
axes[1].set(title='초미세먼지')
axes[1].grid(True, axis='x')
axes[1].set_yticks(range(len(날짜_데이터)))
axes[1].set_yticklabels(날짜_데이터)

plt.suptitle('구로구의 월별 미세먼지와 초미세먼지 농도 비교')
plt.tight_layout()
plt.show()
```

## 구로구의 미세먼지 월별 원그래프 분석
